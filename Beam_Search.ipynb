## Function: `one_point_simulation`

Simulates the logical error rate of a specific quantum code under a given noise rate, a specified circuit and decoder.

```python
def one_point_simulation(
    n: int,
    k: int,
    p_CNOT: float,
    circuit_type: str,
    decoder: str,
    maximum_shots: int = 100_000,
    maximum_errors: int = 100,
)
```

### Allowed Parameter Configurations

The input parameters `(n, k, p_CNOT, circuit_type)` must be selected from the following pre-defined sets:
```
144, 12, 0.001, "normal"
144, 12, 0.002, "normal"
144, 12, 0.003, "normal"
144, 12, 0.004, "normal"
144, 12, 0.005, "normal"
144, 12, 0.006, "normal"

90, 8, 0.002, "normal"
90, 8, 0.003, "normal"
90, 8, 0.004, "normal"
90, 8, 0.005, "normal"
90, 8, 0.002, "XYZ"
90, 8, 0.003, "XYZ"
90, 8, 0.004, "XYZ"
90, 8, 0.005, "XYZ"

450, 32, 0.002, "normal"
450, 32, 0.003, "normal"
450, 32, 0.004, "normal"
450, 32, 0.005, "normal"
```
### Supported Decoders
The `decoder` argument accepts the following values:
* 'bp30+osd'

* 'beam8_230iters'

* 'beam32_340iters'

* 'beam64_640iters'

* 'beam64_32res_640iters'

````python
from simulation_functions import one_point_simulation

one_point_simulation(n=144, k=12, p_CNOT=0.003, circuit_type='normal', decoder='beam8_230iters', maximum_shots=100000, maximum_errors=200)
````

# Function: `decoding_time`

Simulates the runtime of different decoders decoding the [[144,12,12]] BB code with 12 syndrome extraction rounds

```python
def decoding_time(p_CNOT: float, decoder: str, num_shots: int = 10_000)
```

`p_CNOT` allowed values: `0.0005`, `0.001`

`decoder` allowed values: 'bp30+osd', 'beam8_230iters', 'beam32_340iters', 'beam64_640iters'

For every p_CNOT and decoder, this function outputs the average decoding time and 99.9 percentile decoding time

````python
from simulation_functions import decoding_time

decoding_time(p_CNOT=0.0005, decoder="beam8_230iters", num_shots=10000)
````

# Function: `full_simulation`

Runs full simulation for the specified code

```python
def full_simulation(n: int, k: int, maximum_shots: int = 1_000_000, maximum_errors: int = 100)
```

It supports three code instances: [[144, 12, 12]] BB code, [[90, 8, 10]] BB code, and [[450, 32, 8]] HGP code, meaning that the inputs (n, k) must be selected from the following three pairs: (144, 12), (90, 8), and (450, 32)

The simulation results will be stored in a csv file in the `simulation_results` folder

````python
from simulation_functions import full_simulation

full_simulation(n=450, k=32, maximum_shots=100, maximum_errors=100)
````
